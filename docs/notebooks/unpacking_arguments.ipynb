# Unpacking arguments


```
from dewret.renderers.cwl import render
from dewret.tasks import task, construct, workflow

@task()
def f(a: int, b: int,c: int) -> int:
    return a

@workflow()
def foo() -> int:
    return f(**{'a':1, 'b': 2, 'c': 3})

result = foo()

workflow = construct(result, simplify_ids=True)
cwl = render(workflow)
cwl
```

```
from dewret.renderers.cwl import render
from dewret.tasks import task, construct, workflow
from typing import List

@task()
def f(a: int, b: int,c: int) -> List[int]:
    return [a, b, c]

@workflow()
def foo() -> List[int]:
    return f(**{'a':1, 'b': 2, 'c': 3})



result = foo()

workflow = construct(result, simplify_ids=True)
cwl = render(workflow)
cwl
```

```
from dewret.renderers.cwl import render
from dewret.tasks import task, construct, workflow
# from typing import List

@task()
def f(a: int, b: int,c: int) -> list[int]:
    return [a, b, c]

# def g(**Unpack[])

@workflow()
def foo() -> list[int]:
    return f(**{'a':1, 'b': 2, 'c': 3})
    # f(**{'a':1, 'b': 2, 'c': 3}) # doesn't work, because dewret will purge stuff that is not used or returned 


result = foo()

workflow = construct(result, simplify_ids=True)
cwl = render(workflow)
cwl
```

```
from dewret.renderers.cwl import render
from dewret.tasks import task, construct, workflow
from typing import Unpack

class MyArgs(TypedDict):
    a: int
    b: int

def g(**args: Unpack[MyArgs]):
    # Unpack the dictionary into a new dictionary with string keys
    temp: dict[str, int] = dict(args)

    return temp['a']

@workflow()
def foo() -> int:
    return g(**{'a':1, 'b': 2, 'c': 3})

result = foo()

workflow = construct(result, simplify_ids=True)
cwl = render(workflow)
cwl
```
