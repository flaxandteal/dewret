```python
import sys
import yaml
from dewret.tasks import task, construct, workflow
from dewret.renderers.cwl import render
from dewret.annotations import AtRender

# INPUT_PATH: str = "the/input/path"
# OUTPUT_PATH: str = "the/output/path"

INPUT_PATH: AtRender[str] = "the/input/path"
OUTPUT_PATH: AtRender[str] = "the/output/path"

@workflow()
def my_workflow(dummy: str) -> str:
    return f"Read data from {INPUT_PATH} and wrote simulation data to {OUTPUT_PATH}."
    # return f"Read data from {INPUT_PATH} and wrote simulation data to {OUTPUT_PATH}, and also {dummy}"

result = my_workflow(dummy="Hello World")
workflow = construct(result, simplify_ids=True)
cwl = render(workflow)["__root__"]
yaml.dump(cwl, sys.stdout, indent=2)
```
